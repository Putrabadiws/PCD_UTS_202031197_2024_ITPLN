MENGIMPOR LIBRARY

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
from IPython.display import Image

%matplotlib inline
```

MENDEKLARASIKAN GAMBAR

```python
color_image = img.imread('Nama.jpeg')
plt.imshow(color_image)
```

MENDEKLARASIKAN DAN MEMBAGI CITRA WARNA

```python
r = color_image[:, :, 0]
g = color_image[:, :, 1]
b = color_image[:, :, 2]

f, (x1, x2, x3, x4) = plt.subplots(1, 4, figsize = (20,10))

x1.set_title('Citra Kontras')
x1.imshow(color_image)
x1.axis('off')

x2.set_title('Biru')
x2.imshow(b, cmap='gray')
x2.axis('off')

x3.set_title('Merah')
x3.imshow(r, cmap='gray')
x3.axis('off')

x4.set_title('Hijau')
x4.imshow(g, cmap='gray')
x4.axis('off')
```

MENGHITUNG DAN MENDEKLARASIKAN SETIAP WARNA

```python
# Deklarasi warna dalam format RGB
colors = {
    'merah': (255, 0, 0),
    'hijau': (0, 255, 0),
    'biru': (0, 0, 255),
    'kuning': (255, 255, 0),
    'ungu': (128, 0, 128),
    'jingga': (255, 165, 0),
    'putih': (255, 255, 255),
    'hitam': (0, 0, 0),
    # tambahkan warna lain sesuai kebutuhan
}

# Menghitung jumlah warna
jumlah_warna = len(colors)

# Menampilkan daftar warna dan nilai RGB-nya
for warna, rgb in colors.items():
    print(f'Warna {warna}: RGB {rgb}')

print(f"Total warna: {jumlah_warna}")
```

MENAMPILKAN HISTOGRAM DARI MASING-MASING WARNA

```python
plt.subplot(1, 3, 1)
plt.hist(r.ravel(), bins=256, color='red')
plt.title('Histogram Warna Merah')
plt.xlabel('Intensitas')
plt.ylabel('Frekuensi')

# Histogram warna hijau
plt.subplot(1, 3, 2)
plt.hist(g.ravel(), bins=256, color='green')
plt.title('Histogram Warna Hijau')
plt.xlabel('Intensitas')
plt.ylabel('Frekuensi')

# Histogram warna biru
plt.subplot(1, 3, 3)
plt.hist(b.ravel(), bins=256, color='blue')
plt.title('Histogram Warna Biru')
plt.xlabel('Intensitas')
plt.ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

```python
# Membaca gambar
color_image = cv2.imread('Nama.jpeg')

# Konversi citra ke dalam ruang warna HSV
hsv_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2HSV)

# Definisikan rentang warna untuk setiap warna
lower_blue = np.array([90, 100, 100])
upper_blue = np.array([130, 255, 255])

lower_green = np.array([40, 100, 100])
upper_green = np.array([80, 255, 255])

lower_red1 = np.array([0, 100, 100])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([160, 100, 100])
upper_red2 = np.array([180, 255, 255])

# Deteksi warna biru
mask_blue = cv2.inRange(hsv_image, lower_blue, upper_blue)

# Deteksi warna hijau
mask_green = cv2.inRange(hsv_image, lower_green, upper_green)

# Deteksi warna merah
mask_red1 = cv2.inRange(hsv_image, lower_red1, upper_red1)
mask_red2 = cv2.inRange(hsv_image, lower_red2, upper_red2)
mask_red = cv2.bitwise_or(mask_red1, mask_red2)

# Plot hasil
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
plt.title('Original')
#plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(mask_blue, cmap='gray')
plt.title('Blue')
#plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(mask_green, cmap='gray')
plt.title('Green')
#plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(mask_red, cmap='gray')
plt.title('Red')
#plt.axis('off')

plt.tight_layout()
plt.show()
```
